```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training set:
x_train = np.array([1, 2])
y_train = np.array([300, 500])
```

```python
# Model fuction f_wb (return a column array):
def compute_model_prediction_column(x, w, b):
    m = len(x)
    f = np.zeros(m)
    for i in range(m):
        f[i] = w*x[i] + b
        
    return f
```

```python
# Error cost function J (return single number):
def compute_cost(x, y, w, b):
    m = len(x)
    f_wb = np.zeros(m)
    sum = 0
    for i in range(m):
        f_wb[i] = compute_model_prediction_column(x,w,b)[i]
        sum = sum + (f_wb[i] - y[i])**2
    J = (1/2*m) * sum
    
    return J
```

```python
# Plotting error cost as a function of weight w: 
w = np.arange(0, 300, 10)
b = 100

l = len(w)
cost = np.zeros(l) # column array of size l
for i in range(l):
    cost[i] = compute_cost(x_train, y_train, w[i], b)

plt.plot(w, cost)
plt.title("Cost function vs. weight w")
plt.xlabel("w")
plt.ylabel("J(w , b=100)")
plt.show()
```
