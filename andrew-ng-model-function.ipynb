```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training set:

# the input feature (size in 1000 square feet):
x_train = np.array([1, 2])

# the output target (price in $1000):
y_train = np.array([300, 500])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
x_row1 = x_train[0]
y_row1 = y_train[0]
x_row2 = x_train[1]
y_row2 = y_train[1]
print(f"x_row1 = {x_row1}")
print(f"y_row1 = {y_row1}")
print(f"x_row2 = {x_row2}")
print(f"y_row2 = {y_row2}")
```

```python
# Plotting actual values:
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price ($1000)")
plt.show()
```

```python
# Linear regression
# Model fuction f_wb (return a column array):
def compute_model_prediction_column(x, w, b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
        
    return f_wb
```

```python
# Plotting actual values and model predictions:
w = 200
b = 100

y_hat_column = compute_model_prediction_column(x_train, w, b)

plt.scatter(x_train, y_train, marker='x', c='r', label='Actual values')
plt.plot(x_train, y_hat_column, c='b', label='Model predictions')
plt.title("Housing Prices")
plt.xlabel("Size")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# New house size for price prediction:
x_new = [1.2] # in 1000 sqrft
cost_1200_sqft = compute_model_prediction_column(x_new, w, b)
print(f"Price for 1200 sqft is {cost_1200_sqft[0]} thousand dollars")
```
