```python
import numpy as np
import time
```

```python
# scalar shape is ()
# column vector shape is (m,)
# matrix shape is (m,n)

# 1D array of zeros (column vector of zeros):
a = np.zeros(5)
print(a)
print(a.shape)
print(a.dtype)
```

```python
# 1D array of random numbers (column vector of random numbers):
b = np.random.random_sample(5)
print(b)
print(b.shape)
print(b.dtype)
```

```python
# 1D array of equally spaced numbers (column vector):
c = np.arange(5)
print(c)
print(c.shape)
print(c.dtype)
```

```python
# 1D array of user-specified numbers (column vector):
d = np.array([2,5,6,8,9])
print(d)
print(d.shape)
print(d.dtype,"\n")

print(d[0])
print(d[0].shape)
print(d[0].dtype,"\n")

# slicing [start:stop:step]:
print(d[1:3:1])
print(d[1:3:1].shape)
print(d[1:3:1].dtype,"\n")

print(d[:3]) # same as d[0:3]
print(d[3:])
print(d[:])
```

```python
# List of numbers (coma in between):
f = [2,5,7,8]
print(f)
```

```python
# 2D array of zeros (matrix of zeros):
# argument of the zeros function is a tuple (m,n)
a2 = np.zeros((2,4))
print(a2)
print(a2.shape)
print(a2.dtype)
```

```python
# 2D array of random numbers (matrix of random numbers):
b2 = np.random.random_sample((2,4))
print(b2)
print(b2.shape)
print(b2.dtype)
```

```python
# 2D array of user-specified numbers (matrix):
d2 = np.array([[2,5,6,8],[1,2,3,4]])
print(d2)
print(d2.shape)
print(d2.dtype)
```

```python
# dot product by defining a for-loop:
def dot_prod(a,b):
    c = 0
    for i in range(len(a)):
        c = c + a[i]*b[i]
    
    return c
```
